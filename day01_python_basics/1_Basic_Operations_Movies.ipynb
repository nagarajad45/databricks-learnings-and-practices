```python
df = spark.read.table("default.movies")
```

### Displaying the content

```python
display(df)
```

```python
display(df.describe())
```

```python
df.printSchema()
```


### Adding a new column to called Profit (budget * revenue)

```python
df = df.withColumn('profit ', df.revenue * df.budget)
```

```python
display(df)
```


### Renaming the 'revenue' column into 'total_revenue'

```python
df.printSchema
```

```python
df = df.withColumnRenamed("revenue", "total_revenue")
display(df)
```

```python
display(df.show(5))
```

### Column Filtering

```python
df_trimmed = df.select("title", "imdb_rating" ,"industry")
df_trimmed.show(3, truncate=False)
```

### Row Filtering

```python
# Method 1

df_2000_2010 = df.filter((df["release_year"] >= 2000) & (df["release_year"] <= 2010))
df_2000_2010.show(3, truncate=False)
```

```python
# Method 2
from pyspark.sql.functions import col

df_trimmed = df.select("title", "release_year", "imdb_rating")
df_filtered = df.filter(
    col("imdb_rating") >= 9 
)

df_filtered.show(3, truncate = False)

```

```python
print(f"no. of movies with >= 9: {df_filtered.count()}")
display(df_filtered)
```

```python
# Method 3

df_filtered= df.filter(col("release_year").between(2000, 2010))
df_filtered.show(3, truncate=False)
```

```python
# print title of the movies and studio which are released by Marvel Studio

df_filtered = df.filter(df["studio"] == "Marvel Studios")
df_filtered.show(3, truncate=False)
```

```python
# Check the what are the unique industries are there in industry column

unique_industries = df.select("industry").distinct()
display(unique_industries)
```

```python
# Check the what are the unique languages are there in language column

unique_languages = df.select("language").distinct()
display(unique_languages)
```
